```python
def create_dataset( dataset, time_diff = 1 ):
    dataX, dataY = [], []
    for i in range( len( dataset ) - time_diff - 1 ):
        a = dataset[ i:( i + time_diff ), 0 ]
        dataX.append( a )
        dataY.append( dataset[ i + time_diff, 0 ] )
    return np.array( dataX ), np.array( dataY )
```

```python
time_diff = 1

train_X, train_Y = create_dataset( train, time_diff )
test_X, train_Y = create_dataset( test, time_diff )
```
