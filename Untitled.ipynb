```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose

from matplotlib import pyplot as plt

import warnings
```

```python
df_bitcoin = pd.read_csv('btc-cleaned.csv')

# Go through each column and print amount of NaN values 
print(df_bitcoin.isna().sum())

# Print first ( no. of rows ) -> ( 10 )
print(df_bitcoin.head(10))
```

```python
# Estamate value between two known points from dataset
df_bitcoin.interpolate();

df_bitcoin.shape

plt.figure( figsize = ( 15 , 5 ) )

df_bitcoin['PriceUSD'].plot( title='Cryptocurrency Price Analysis - BITCOIN' )

plt.legend(['Price of Bitcoin over time'], loc = 'upper left')

plt.ylabel('Price in $ | USD ' + '(PriceUSD)')
plt.xlabel('time' + ' (t) ' + 'interval in days')
```

```python
df_bitcoin['PriceUSD']=df_bitcoin['PriceUSD'].replace(np.nan,df_bitcoin['PriceUSD'].mean())
df_bitcoin['time']=df_bitcoin['time'].replace(np.nan,df_bitcoin['time'].mode()[0])
```

```python
df_bitcoin.drop(df_bitcoin.columns.difference(['time','PriceUSD']), 1, inplace=True)
df_bitcoin.drop(['time'], 1, inplace=True)
```

```python
prediction_days = 30

df_bitcoin['PREDICTION'] = df_bitcoin[['PriceUSD']].shift(-prediction_days)
```

```python
X = np.array(df_bitcoin.drop(['PREDICTION'], 1))
X = X[:len(df_bitcoin)-prediction_days]
print(X)

y = np.array(df_bitcoin['PREDICTION'])
y = y[:-prediction_days]
print(y)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

prediction_days = np.array(df_bitcoin.drop(['PREDICTION'], 1))[-prediction_days:]
print(prediction_days)
```

```python
from sklearn.svm import SVR 

svr_rbf = SVR(kernel='rbf', C=1e3, gamma = 0.00001)

svr_rbf.fit(x_train, y_train)

svr_rbf_score = svr_rbf.score(x_test, y_test)

print(svr_rbf_score)

#pipeline = Pipeline([('scaler', StandardScaler()), ('svc', SVC())])

#pipeline.fit(x_train, y_train)

#pipeline.score(x_test, y_test)

print(df_bitcoin)
```

```python
svm_prediction = svr_rbf.predict(x_test)

print(svm_prediction)

print()

print(y_test)
```

```python
svm_prediction = svr_rbf.predict(prediction_days)

print(svm_prediction)

#print()

#print(df_bitcoin.tail(prediction_days))
```

```python
def mean_absolute_percentage_error(x_test, svm_prediction): 
    X_test, test_predictions = np.array(x_test), np.array(svm_prediction)
    return np.mean(np.abs((x_test - test_predictions) / svm_prediction)) * 100
mean_absolute_percentage_error(x_test, svm_prediction)
```

```python
precision = precision_score( y_test, svm_prediction )
recall = recall_score( y_test, svm_prediction )

print( 'Precision : ', precision )
print( 'Recall : ', recall )
```
